```python
x=df.drop('class',axis=1)
y=df['class']
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.30, random_state=1)
```

KNeighborsClassifier

```python
from sklearn.neighbors import KNeighborsClassifier 
```

```python
KNN=KNeighborsClassifier(n_neighbors=5, weights='distance')
```

```python
KNN.fit(x_train,y_train)
```

```python
predicted_1=KNN.predict(x_test)
```

```python
predicted_1
```

```python
from scipy.stats import zscore
print('KNN predicting accuracy {0:.2g}%'.format(KNN.score(x_test,y_test)*100))
```

```python
from sklearn.svm import SVC
svc=SVC()
svc.fit(x_train,y_train)
```

```python
predicted_2=svc.predict(x_test)
predicted_2
```

```python
print('SVM predicting accuracy {0:.2g}%'.format(svc.score(x_test,y_test)*100))
```

```python
knnpred=pd.DataFrame(predicted_1)
svcpred=pd.DataFrame(predicted_2) 

df1=pd.concat([knnpred,svcpred],axis=1)
```
